```python
import os
import shutil
import random
from glob import glob
from ultralytics import YOLO
import cv2

# === ConfiguraciÃ³n del entorno ===
ruta_modelo = "/content/drive/MyDrive/Colab Notebooks/personal-tests/tesis/resultados-scratch/yolo11/results_train_fold6/fold6/weights/best.pt"
ruta_test = "/content/drive/MyDrive/Colab Notebooks/personal-tests/tesis/data/merged/images"
ruta_resultados = "/content/drive/MyDrive/Colab Notebooks/personal-tests/tesis/resultados-tesis-test"
num_imagenes = 100
umbral_confianza = 0.6  # Umbral del modelo seleccionado

# === Crear carpeta de resultados ===
if os.path.exists(ruta_resultados):
    shutil.rmtree(ruta_resultados)
os.makedirs(ruta_resultados, exist_ok=True)

# === Cargar modelo YOLO ===
modelo = YOLO(ruta_modelo)

# === Seleccionar imÃ¡genes aleatorias ===
extensiones = ["*.jpg", "*.jpeg", "*.png"]
imagenes_test = []
for ext in extensiones:
    imagenes_test.extend(glob(os.path.join(ruta_test, ext)))
imagenes_seleccionadas = random.sample(imagenes_test, num_imagenes)

# === Inferencia con umbral y visualizaciÃ³n personalizada ===
for img_path in imagenes_seleccionadas:
    nombre_archivo = os.path.basename(img_path)

    # Leer la imagen original
    img = cv2.imread(img_path)

    # Realizar predicciÃ³n
    resultados = modelo(img_path, conf=umbral_confianza, iou=0.7)

    for r in resultados:
        boxes = r.boxes
        for box in boxes:
            x1, y1, x2, y2 = map(int, box.xyxy[0].tolist())
            conf = float(box.conf[0])
            if conf < umbral_confianza:
                continue

            # Dibujar caja en verde (BGR) con grosor 3
            cv2.rectangle(img, (x1, y1), (x2, y2), color=(0, 255, 0), thickness=3)

            # Agregar texto: "C: <confianza>"
            label = f"C: {conf:.2f}"
            cv2.putText(img, label, (x1, y1 - 10), cv2.FONT_HERSHEY_SIMPLEX,
                        0.8, (0, 255, 0), 2)

    # Guardar imagen con resultados
    cv2.imwrite(os.path.join(ruta_resultados, nombre_archivo), img)

print("ImÃ¡genes procesadas y guardadas en la carpeta:", ruta_resultados)
```
