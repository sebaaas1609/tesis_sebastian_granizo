```python
model_path = ROOT_DIR + '/yolov10m.pt'
model_10 = YOLO(model_path)

# Train the model
model_10.train(data=os.path.join(ROOT_DIR, 'drive_config.yaml'), epochs=1000, imgsz=640, batch=32, patience=50, optimizer="AdamW",
               project=yolo10_project_dir, freeze=10, plots=True, hsv_h=0.015, hsv_s=0.7, hsv_v=0.4,
                degrees=45.0, translate=0.1, scale=0.5, shear=0.1,
                fliplr=0.5, flipud=0.3, mixup=0.2, mosaic=1.0)
```

```python
model_path_12 = ROOT_DIR + '/yolo12s.pt'
yolo12v2_project_dir = ROOT_DIR + '/yolo12v2'
model_12_v2 = YOLO(model_path_12)


model_12_v2.train(
    data=os.path.join(ROOT_DIR, 'train_config.yaml'),
    epochs=1000,
    imgsz=640,
    patience=50,
    optimizer="AdamW",
    project=yolo12v2_project_dir,
    freeze=10,
    plots=True,

    # Learning rate and optimizer settings
    lr0=0.002, lrf=0.01, cos_lr=True,

    # Warmup settings
    warmup_epochs=5, warmup_momentum=0.9, warmup_bias_lr=0.2,

    # Augmentations
    hsv_h=0.015, hsv_s=0.7, hsv_v=0.4,
    degrees=45.0, translate=0.1, scale=0.5, shear=0.1,
    fliplr=0.5, flipud=0.3, mixup=0.2, mosaic=1.0,

    # Batch and accumulation
    batch=32,

    # Regularization and smoothing
    weight_decay=0.0001, label_smoothing=0.1,

    # Loss adjustments
    box=8.0, cls=0.6, dfl=1.5,

    # Precision and multi-scale training
    amp=True, multi_scale=True
)

```
